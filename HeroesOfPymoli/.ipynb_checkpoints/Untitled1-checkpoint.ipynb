```python
# dependencies
import pandas as pd


# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read csv into DF
purchase_data_df = pd.read_csv(csv_path)
purchase_data_df.head()
```

```python
# Display the total number of players

players = purchase_data_df.groupby ("SN")["SN"].nunique()
players.count() 
```

```python
# Total Number of purchasing analysis
average_purchase_price = purchase_data_df["Price"].mean()
total_revenue = purchase_data_df["Price"].sum()
number_purchase = purchase_data_df["Price"].count()
unique_items = len(purchase_data_df["Item ID"].unique())

# Create a summary DataFrame to hold results
purchasing_analysis = pd.DataFrame({"Number of Unique Items": unique_items,
                               "Total Revenue": [total_revenue],
                               "Number of Purchases": [number_purchase],
                               "Average Purchase Price": [average_purchase_price]})

# Display data cleaner formatting
purchasing_analysis = purchasing_analysis.round(2)
purchasing_analysis ["Average Price"] = purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis ["Total Revenue"] = purchasing_analysis["Total Revenue"].map("${:,.2f}".format)
purchasing_analysis = purchasing_analysis.loc[:, ["Number of Unique Items", "Average Purchase Price", "Number of Purchases", "Total Revenue"]]

# Display the summary DataFrame
purchasing_analysis
```

```python
# Percentage and count by gender demographics
gender_demographics = purchase_data_df.groupby("Gender").agg({"SN":"unique"})
gender_demographics ["Total Count"] = gender_demographics["SN"].apply(len)
gender_demographics .drop(columns="SN", inplace=True)

gender_demographics["Percentage of Players"] = gender_demographics["Total Count"] / gender_demographics["Total Count"].sum() * 100
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].map("{:.2f}%".format)
gender_demographics.sort_values("Total Count", ascending=False,inplace=True)
gender_demographics

```

```python
# Purchasing analysis by gender
gender_purchase_total = purchase_data_df.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_average = purchase_data_df.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Value")
gender_counts = purchase_data_df.groupby(["Gender"]).count()["Price"].rename("Purchase Counts")

avg_total_purchase_per_person = gender_purchase_total / gender_demographics["Total Count"]
                                                             
gender_analysis_data = pd.DataFrame({"Purchase Count":gender_counts,
                            "Average Purchase Value": gender_average,
                            "Total Purchase Value": gender_purchase_total,
                            "Average Total Purchase Per Person":avg_total_purchase_per_person})
gender_analysis_data
```
